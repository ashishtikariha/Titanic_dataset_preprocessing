## apply() : It takes function as argument & applies that function to all the values of a series

```python
# Create one more column 'sex_1' with 1 for male & 0 for female
```

```python
df['sex_1'] = df.Sex.apply(lambda x : 1 if x == 'male' else 0)
df
```

```python
df['sex_2'] = df.Sex.apply(lambda x : 'M' if x == 'male' else 'F')
```

```python
df
```

```python
df['sex_3'] = df.Sex.replace(['male' , 'female'] ,[1 , 0])
df
```

```python
# age : all nan values impute it with median
df.Age.fillna(df.Age.median()) 
df
```
